# Basic Example

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

## Basic Matplotlib Commands

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

## Creating Multiplots on Same Canvas

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

## Setting colors, linewidths, linetypes

### Colors with MatLab like syntax

```python
# MATLAB style line color and style 
plt.subplots()
plt.plot(x, x**2, 'b.-') # blue line with dots
plt.plot(x, x**3, 'g--') # green dashed line
```

### Colors with the color= parameter

```python
plt.subplots()

plt.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
plt.plot(x, x+2, color="#8B008B")        # RGB hex code
plt.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```
